```python
import numpy as np
import matplotlib.pyplot as plt
matrix = 35.8 # Физический размер матрицы
resolution = 3840 # Максимальное разрешение
Sample_period = round((matrix/ resolution), 4) # Период дискретизации
Sample_rate = round(1/Sample_period, 1) # Частота дискретизации
fs = 3 * Sample_rate # FS
n_px = 15 # Выходной размер изображения
x = np.arange(0, Sample_period * n_px, Sample_period / 100) # X
y=0.5 + 0.5*np.sin(fs * x ) # Y
plt.plot(x, y) # Отрисовка
plt.show()
```
